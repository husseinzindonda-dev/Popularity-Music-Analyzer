```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')

class SpotifyAnalyzer:
    def __init__(self, file_path):
        self.file_path = file_path
        self.df = None
        self.numerical_features = None
        self.correlation_matrix = None
        
    def load_and_explore_data(self):
        """Load dataset and perform initial exploration"""
        try:
            self.df = pd.read_csv(self.file_path)
            print("‚úÖ Dataset loaded successfully!")
        except FileNotFoundError:
            print(f"‚ùå File {self.file_path} not found!")
            return False
        
        # Comprehensive data exploration
        self._display_dataset_info()
        self._check_data_quality()
        return True
    
    def _display_dataset_info(self):
        """Display comprehensive dataset information"""
        print("\n" + "="*50)
        print("üìä DATASET OVERVIEW")
        print("="*50)
        
        print(f"Shape: {self.df.shape}")
        print(f"Memory usage: {self.df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
        
        print("\nüìù Column Information:")
        print(f"Total columns: {len(self.df.columns)}")
        for i, col in enumerate(self.df.columns, 1):
            print(f"  {i:2d}. {col}")
        
        print("\nüîç Data Types:")
        print(self.df.dtypes)
        
        print("\nüìã First 5 rows:")
        print(self.df.head())
        
        print("\nüìà Basic Statistics:")
        print(self.df.describe())
    
    def _check_data_quality(self):
        """Check data quality and identify issues"""
        print("\n" + "="*50)
        print("üîç DATA QUALITY CHECK")
        print("="*50)
        
        # Missing values
        missing_values = self.df.isnull().sum()
        if missing_values.sum() > 0:
            print("‚ùå Missing values found:")
            print(missing_values[missing_values > 0])
        else:
            print("‚úÖ No missing values")
        
        # Duplicates
        duplicates = self.df.duplicated().sum()
        if duplicates > 0:
            print(f"‚ùå {duplicates} duplicate rows found")
        else:
            print("‚úÖ No duplicate rows")
        
        # Identify numerical features
        self.numerical_features = self.df.select_dtypes(include=[np.number]).columns.tolist()
        print(f"\nüî¢ Numerical features: {self.numerical_features}")
    
    def analyze_correlations(self):
        """Perform correlation analysis with visualization"""
        if self.numerical_features is None:
            print("‚ùå Please load data first!")
            return
        
        print("\n" + "="*50)
        print("üìà CORRELATION ANALYSIS")
        print("="*50)
        
        # Calculate correlation matrix
        self.correlation_matrix = self.df[self.numerical_features].corr()
        
        print("Correlation Matrix:")
        print(self.correlation_matrix.round(3))
        
        # Find strongest correlations
        self._find_strong_correlations()
        
        # Visualize correlations
        self._plot_correlation_heatmap()
        self._plot_correlation_clustermap()
    
    def _find_strong_correlations(self):
        """Identify strongest positive and negative correlations"""
        corr_matrix = self.correlation_matrix.copy()
        np.fill_diagonal(corr_matrix.values, np.nan)  # Ignore diagonal
        
        # Flatten correlation matrix and find extremes
        corr_pairs = corr_matrix.unstack().sort_values(ascending=False)
        
        print("\nüî• Strongest Positive Correlations:")
        print(corr_pairs.head(6))
        
        print("\nüßä Strongest Negative Correlations:")
        print(corr_pairs.tail(6))
    
    def _plot_correlation_heatmap(self):
        """Create a heatmap of correlations"""
        plt.figure(figsize=(12, 10))
        
        mask = np.triu(np.ones_like(self.correlation_matrix, dtype=bool))
        
        sns.heatmap(self.correlation_matrix, 
                   mask=mask,
                   annot=True, 
                   cmap='coolwarm', 
                   center=0,
                   square=True,
                   fmt='.3f',
                   cbar_kws={"shrink": .8})
        
        plt.title('Spotify Features Correlation Matrix\n(Triangle Masked)', fontsize=16, pad=20)
        plt.xticks(rotation=45, ha='right')
        plt.yticks(rotation=0)
        plt.tight_layout()
        plt.show()
    
    def _plot_correlation_clustermap(self):
        """Create a clustered heatmap"""
        plt.figure(figsize=(12, 10))
        
        g = sns.clustermap(self.correlation_matrix, 
                          annot=True, 
                          cmap='coolwarm', 
                          center=0,
                          fmt='.2f',
                          figsize=(12, 10))
        
        g.fig.suptitle('Clustered Correlation Matrix of Spotify Features', 
                      fontsize=16, y=1.02)
        plt.show()
    
    def perform_pca_analysis(self):
        """Perform Principal Component Analysis"""
        from sklearn.decomposition import PCA
        
        print("\n" + "="*50)
        print("üîÆ PRINCIPAL COMPONENT ANALYSIS")
        print("="*50)
        
        # Select numerical features and standardize
        X = self.df[self.numerical_features]
        X_standardized = StandardScaler().fit_transform(X)
        
        # Perform PCA
        pca = PCA()
        principal_components = pca.fit_transform(X_standardized)
        
        # Explained variance
        explained_variance = pca.explained_variance_ratio_
        cumulative_variance = np.cumsum(explained_variance)
        
        print("Explained Variance Ratio:")
        for i, (var, cum_var) in enumerate(zip(explained_variance, cumulative_variance)):
            print(f"PC{i+1}: {var:.3f} ({cum_var:.3f} cumulative)")
        
        # Plot explained variance
        self._plot_variance_explained(explained_variance, cumulative_variance)
        
        # Create PCA dataframe
        pca_df = pd.DataFrame(data=principal_components[:, :2], 
                             columns=['PC1', 'PC2'])
        pca_df['Artist'] = self.df['Artist']
        
        # Plot PCA results
        self._plot_pca_results(pca_df, pca, X.columns)
        
        return pca, principal_components
    
    def _plot_variance_explained(self, explained_variance, cumulative_variance):
        """Plot PCA explained variance"""
        plt.figure(figsize=(12, 6))
        
        plt.subplot(1, 2, 1)
        plt.bar(range(1, len(explained_variance) + 1), explained_variance, alpha=0.6)
        plt.xlabel('Principal Components')
        plt.ylabel('Explained Variance Ratio')
        plt.title('Individual Explained Variance')
        
        plt.subplot(1, 2, 2)
        plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, 'bo-')
        plt.xlabel('Number of Principal Components')
        plt.ylabel('Cumulative Explained Variance')
        plt.title('Cumulative Explained Variance')
        plt.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.show()
    
    def _plot_pca_results(self, pca_df, pca, feature_names):
        """Plot PCA results with artist labels"""
        plt.figure(figsize=(14, 10))
        
        # Scatter plot
        scatter = plt.scatter(pca_df['PC1'], pca_df['PC2'], 
                             c=range(len(pca_df)), cmap='viridis', s=100)
        
        # Add artist labels
        for i, artist in enumerate(pca_df['Artist']):
            plt.annotate(artist, (pca_df['PC1'][i], pca_df['PC2'][i]),
                        xytext=(5, 5), textcoords='offset points',
                        fontsize=8, alpha=0.7)
        
        plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.2%} variance)')
        plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.2%} variance)')
        plt.title('PCA: Spotify Songs in 2D Space\n(Colored by Artist)', fontsize=16)
        plt.colorbar(scatter, label='Song Index')
        plt.grid(True, alpha=0.3)
        plt.show()
    
    def generate_report(self):
        """Generate comprehensive analysis report"""
        print("\n" + "="*50)
        print("üìã COMPREHENSIVE ANALYSIS REPORT")
        print("="*50)
        
        # Most popular artist
        avg_popularity = self.df.groupby('Artist')['Popularity (/100)'].mean().sort_values(ascending=False)
        print(f"\nüé§ Most Popular Artist: {avg_popularity.index[0]} ({avg_popularity.iloc[0]:.1f}/100)")
        
        # Most streamed song
        most_streamed = self.df.loc[self.df['Streams (billions)'].idxmax()]
        print(f"üî• Most Streamed: '{most_streamed['Track Name']}' by {most_streamed['Artist']} ({most_streamed['Streams (billions)']}B streams)")
        
        # Feature ranges
        print(f"\nüìä Feature Ranges:")
        for feature in ['Tempo (BPM)', 'Danceability (/100)', 'Energy (/100)']:
            min_val = self.df[feature].min()
            max_val = self.df[feature].max()
            print(f"  {feature}: {min_val} - {max_val}")

# Main execution
if __name__ == "__main__":
    # Initialize analyzer
    analyzer = SpotifyAnalyzer('spotify_top_20_analysis.csv')
    
    # Perform complete analysis
    if analyzer.load_and_explore_data():
        analyzer.analyze_correlations()
        analyzer.perform_pca_analysis()
        analyzer.generate_report()
```
